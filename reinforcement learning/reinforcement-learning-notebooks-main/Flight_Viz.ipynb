### Visualization Plot

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
df_flight_paths = pd.read_csv('https://raw.githubusercontent.com/DRA-chaos/QuantumClimax/main/Data%20Folder/flights_modified.csv?token=GHSAT0AAAAAABPR6P2X4FMXZ46YGAY3Y5UQYSC3WQQ')
df_flight_paths.head()
```

```python
fig = go.Figure()

# fig.add_trace(go.Scattergeo(
#     locationmode = 'country-names',
#     lon = df_flight_paths['start_longitudinal'],
#     lat = df_flight_paths['end_longitudinal'],
#     hoverinfo = 'text',
#     mode = 'markers',
#     marker = dict(
#         size = 2,
#         color = 'rgb(255, 0, 0)',
#         line = dict(
#             width = 3,
#             color = 'rgba(68, 68, 68, 0)'
#         )
#     )))

lons = []
lats = []
lons = np.empty(3 * len(df_flight_paths))
lons[::3] = df_flight_paths['start_longitudinal']
lons[1::3] = df_flight_paths['end_longitudinal']
lons[2::3] = None
lats = np.empty(3 * len(df_flight_paths))
lats[::3] = df_flight_paths['start_latitudinal']
lats[1::3] = df_flight_paths['end_latitudinal']
lats[2::3] = None

fig.add_trace(
    go.Scattergeo(
        locationmode = 'country names',
        lon = lons,
        lat = lats,
        mode = 'lines',
        line = dict(width = 1,color = 'red'),
        opacity = 0.5
    )
)

fig.update_layout(
    title_text = 'European Airlines',
    showlegend = False,
    geo = go.layout.Geo(
        scope = 'europe',
        projection_type = 'azimuthal equal area',
        showland = True,
        landcolor = 'rgb(243, 243, 243)',
        countrycolor = 'rgb(204, 204, 204)',
    ),
    height=700,
)

fig.show()
```
