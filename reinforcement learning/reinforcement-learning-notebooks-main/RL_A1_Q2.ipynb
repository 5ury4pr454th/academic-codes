```python
# Initialize some variables

policy = {}
rewards = {}
gamma = 0.9
eps = 0.0001
all_states = ['hostel', 'canteen', 'academic building']
possible_actions = ['eat', 'attend']
```

```python
# since rewards are independent of actions and previous states, and only dependent on next states, we have r(s'|s,a) = r(s')

rewards['hostel'] = -1
rewards['canteen'] = 1
rewards['academic building'] = 3
```

```python
# assuming the following indexes
# hostel = 0, canteen = 1, academic building = 2
# eat = 0, attend = 1

probability_matrix = np.zeros((len(all_states), len(all_states), len(possible_actions)))

# now, create a probability_matrix (in this case of size 3x3x2 (initial_state x final_state x actions))

## for action = eat
probability_matrix[0, 1, 0] = 1

probability_matrix[1, 1, 0] = 1

probability_matrix[2, 1, 0] = 0.8
probability_matrix[2, 2, 0] = 0.2

## for action = attend
probability_matrix[0, 0, 1] = 0.5
probability_matrix[0, 2, 1] = 0.5

probability_matrix[1, 0, 1] = 0.3
probability_matrix[1, 1, 1] = 0.1
probability_matrix[1, 2, 1] = 0.6

probability_matrix[2, 0, 1] = 0.3
probability_matrix[2, 2, 1] = 0.7
```

## Policy Iteration

```python
# assume that the value function for all states is 0

v = {}

v['hostel'] = 0
v['canteen'] = 0
v['academic building'] = 0

print(v)
```

```python
# initialize a random policy

for state in all_states:
  policy[state] = np.random.choice(possible_actions)

print(policy)
```

```python
# a mapping

state_index = {}
state_index['hostel'] = 0
state_index['canteen'] = 1
state_index['academic building'] = 2

action_index = {}
action_index['eat'] = 0
action_index['attend'] = 1
```

```python
# evaluate policy

def policy_eval():
  
  delta = 1.00
  
  while delta > eps:
    delta = 0.00

    for each_state in all_states:
      v0 = v[each_state]
      value_sum = 0.00
      
      for state in all_states:
        # since r(s'|s,a) = r(s'), we have p(s',r|s,pi(s)) = p(s'| s, pi(s))
        value_sum += probability_matrix[state_index[each_state], state_index[state], action_index[policy[each_state]]]*(rewards[state] + gamma*v[state])
      v[each_state] = value_sum
      delta = max(delta, np.abs(v0 - value_sum))
```

```python
# update policy

def update_policy():
  
  stable_policy = True
  
  for each_state in all_states:

    a = policy[each_state]
      
    value_sum_for_action = {}
    value_sum_for_action['eat'] = 0
    value_sum_for_action['attend'] = 0

    for state in all_states:
      # since r(s'|s,a) = r(s'), we have p(s',r|s,pi(s)) = p(s'| s, pi(s))
      value_sum_for_action['eat'] += probability_matrix[state_index[each_state], state_index[state], action_index['eat']]*(rewards[state] + gamma*v[state])
      value_sum_for_action['attend'] += probability_matrix[state_index[each_state], state_index[state], action_index['attend']]*(rewards[state] + gamma*v[state])
    
    if value_sum_for_action['eat'] > value_sum_for_action['attend']:
      policy[each_state] = 'eat'
    else:
      policy[each_state] = 'attend'
    
    if a!= policy[each_state]:
      stable_policy = False
```

```python
def run_policy_iteration():
  
  stable_policy = True

  policy_eval()
  
  update_policy()

  while not stable_policy:

    policy_eval()
    
    update_policy()
```

```python
run_policy_iteration()
```

## Value Iteration

```python
# assume that the value function for all states is 0

v = {}

v['hostel'] = 0
v['canteen'] = 0
v['academic building'] = 0

print(v)
```

```python
deterministic_policy = {}

for state in all_states:
  deterministic_policy[state] = np.random.choice(possible_actions)

print(deterministic_policy)
```

```python
def run_value_iteration():
  
  delta = 1.00

  while delta > eps:
    
    delta = 0.00

    for each_state in all_states:
      
      v0 = v[each_state]
      a = policy[each_state]
        
      value_sum_for_action = {}
      value_sum_for_action['eat'] = 0
      value_sum_for_action['attend'] = 0
      max_value_sum = 0

      for state in all_states:

        # since r(s'|s,a) = r(s'), we have p(s',r|s,pi(s)) = p(s'| s, pi(s))
        
        value_sum_for_action['eat'] += probability_matrix[state_index[each_state], state_index[state], action_index['eat']]*(rewards[state] + gamma*v[state])
        value_sum_for_action['attend'] += probability_matrix[state_index[each_state], state_index[state], action_index['attend']]*(rewards[state] + gamma*v[state])

      max_value_sum = max(value_sum_for_action['eat'], value_sum_for_action['attend'])
      v[each_state] = max_value_sum
      delta = max(delta, np.abs(v0 - max_value_sum))


  ### getting the deterministic policy

  for each_state in all_states:
    
    value_sum_for_action = {}
    value_sum_for_action['eat'] = 0
    value_sum_for_action['attend'] = 0


    for state in all_states:

      # since r(s'|s,a) = r(s'), we have p(s',r|s,pi(s)) = p(s'| s, pi(s))  
      value_sum_for_action['eat'] += probability_matrix[state_index[each_state], state_index[state], action_index['eat']]*(rewards[state] + gamma*v[state])
      value_sum_for_action['attend'] += probability_matrix[state_index[each_state], state_index[state], action_index['attend']]*(rewards[state] + gamma*v[state])

    if value_sum_for_action['eat'] > value_sum_for_action['attend']:
      deterministic_policy[each_state] = 'eat'
    else:
      deterministic_policy[each_state] = 'attend'

  
```

```python
print("Optimal value for each state is ", v)
print("Optimal policy is ", deterministic_policy)
```

```python
def compare_iterations():
  
  v = {}
  policy = {}
  deterministic_policy = {}

  v['hostel'] = 0
  v['canteen'] = 0
  v['academic building'] = 0

  for state in all_states:
    deterministic_policy[state] = np.random.choice(possible_actions)
    policy[state] = np.random.choice(possible_actions)

  ### performing policy iteration
  run_policy_iteration()
  print("Optimal policy from policy iteration:\n", policy)
  print() 

  v['hostel'] = 0
  v['canteen'] = 0
  v['academic building'] = 0

  run_value_iteration()
  print("Optimal value for each state from value iteration\n", v)
  print("Optimal policy for each state from value iteration\n", deterministic_policy)
```
