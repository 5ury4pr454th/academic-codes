```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
og_img2 = cv.imread("/content/Fig2.tif",cv.IMREAD_GRAYSCALE)
# get threshold by otsu
(thresh, binary_img) = cv.threshold(og_img2, 128, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
# find all contours
contours, hierarchy = cv.findContours(binary_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
og_img2_copy = cv.imread("/content/Fig2.tif")
# draw contours
final_img = cv.drawContours(image=og_img2_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv.LINE_4)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 20))
ax[0].imshow(og_img2, cmap = "gray")
ax[1].imshow(og_img2_copy)
```
