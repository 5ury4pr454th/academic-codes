```python
import numpy as np
import scipy as scp
import cv2 as cv
import os

from google.colab.patches import cv2_imshow
```

```python
img = cv.imread("/content/uhtAZT.jpg")
cv2_imshow(img)
```

```python
low_sigma_img = cv.GaussianBlur(img, (3, 3), 0)
high_sigma_img = cv.GaussianBlur(img, (5, 5), 0)
DoG = low_sigma_img - high_sigma_img

def dog_algo(img, a, b):
  low_sigma_img = cv.GaussianBlur(img, (a, a), 0)
  high_sigma_img = cv.GaussianBlur(img, (b, b), 0)
  return cv2_imshow(low_sigma_img - high_sigma_img)
```

```python
dog_algo(img, 3, 5)
```

```python
dog_algo(img, 3, 9)
```

```python
dog_algo(img, 9, 15)
```

```python
dog_algo(img, 9, 11)
```

```python
from matplotlib import pyplot as plt
import numpy as np

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))/(sig * np.power(2*np.pi, 0.5))

x = np.linspace(-5, 5, 400)
plt.plot(x, gaussian(x, 0, 1))
plt.plot(x, gaussian(x, 0, 4))
```
